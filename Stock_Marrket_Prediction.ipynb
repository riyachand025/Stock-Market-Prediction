### Preprocessing both datasets

# Preprocessing news headlines dataset
1. Converting date to proper datetime objects.
2. Removing duplicate entries.
3. Grouping headlines based on their publish date.
4. Setting date as index and sorting dataset based on it.

```python
headlines['publish_date'] = pd.to_datetime(headlines['publish_date'], format='%Y%m%d')
headlines = headlines.drop_duplicates()
```

```python
headlines = headlines.groupby('publish_date').agg({
    'headline_text': lambda x: ' '.join(x)
}).reset_index()
```

```python
headlines.set_index('publish_date', inplace=True)
headlines.sort_index(inplace=True)
```

### Textual Data Analysis

Combining the sentiments of the headlines

```python
stock_dataset['pos'] = None
stock_dataset['neg'] = None
stock_dataset['comp'] = None
stock_dataset['neu'] = None

stock_dataset.head()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
    if pd.isna(text):
        return {'pos': None, 'neg': None, 'compound': None, 'neu': None}
    sentiment = sid.polarity_scores(text)
    return sentiment
```

```python
sentiment_scores = stock_dataset['headline_text'].apply(lambda text: analyze_sentiment(text))
stock_dataset[['pos', 'neg', 'comp', 'neu']] = pd.DataFrame(sentiment_scores.tolist(), index=stock_dataset.index)
```

## Creating Shifted Dataset

```python
start_date = pd.to_datetime('2001-01-02')
end_date = pd.to_datetime('2022-03-30')
```

```python
# Extract columns using datetime index range
close_price = stock_dataset.loc[start_date:end_date, 'Close']
compound = stock_dataset.loc[start_date:end_date, 'comp']
negative = stock_dataset.loc[start_date:end_date, 'neg']
neutral = stock_dataset.loc[start_date:end_date, 'neu']
positive = stock_dataset.loc[start_date:end_date, 'pos']
open_price = stock_dataset.loc[start_date:end_date, 'Open']
high = stock_dataset.loc[start_date:end_date, 'High']
low = stock_dataset.loc[start_date:end_date, 'Low']
volume = stock_dataset.loc[start_date:end_date, 'Volume']
```

```python
close_price
```

Shifting the close price and comp

```python
close_price_shifted = stock_dataset['Close'].shift(-1)
compound_shifted = stock_dataset['comp'].shift(-1)
```

```python
shifted_dataset = pd.DataFrame({
    'close_price': stock_dataset['Close'],
    'close_price_shifted': close_price_shifted,
    'compound': stock_dataset['comp'],
    'compound_shifted': compound_shifted,
    'volume': stock_dataset['Volume'],
    'open_price': stock_dataset['Open'],
    'high': stock_dataset['High'],
    'low': stock_dataset['Low']
})
```

```python
shifted_dataset.tail()
```

```python
shifted_dataset = shifted_dataset.dropna()
```

Creating training data and distributing it

```python
X = shifted_dataset[['close_price', 'compound', 'compound_shifted', 'volume', 'open_price', 'high', 'low']]
y = shifted_dataset['close_price_shifted']
```

```python
X
```

```python
y
```

### Scaling Data using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
```

```python
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1)).ravel()
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
y_scaled_df = pd.Series(y_scaled, name='close_price_shifted', index=X.index)
```

```python
X_scaled_df
```

```python
y_scaled_df
```

### Splitting Dataset into training, validation and test dataset

```python
from sklearn.model_selection import train_test_split

X_train_val, X_test, y_train_val, y_test = train_test_split(X_scaled_df, y_scaled_df, test_size=0.2, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.25, shuffle=False)
```

```python
print(f"Training Features (X_train) shape: {X_train.shape}")
print(f"Validation Features (X_val) shape: {X_val.shape}")
print(f"Test Features (X_test) shape: {X_test.shape}")
print(f"Training Target (y_train) shape: {y_train.shape}")
print(f"Validation Target (y_val) shape: {y_val.shape}")
print(f"Test Target (y_test) shape: {y_test.shape}")
```

### Creating Sequential Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, Dense, Activation
```

```python
np.random.seed(2)

columns = ['close_price', 'compound', 'compound_shifted', 'volume', 'open_price', 'high', 'low']

model=Sequential()
model.add(LSTM(100,return_sequences=True,activation='tanh',input_shape=(len(columns),1)))
model.add(Dropout(0.1))
model.add(LSTM(100,return_sequences=True,activation='tanh'))
model.add(Dropout(0.1))
model.add(LSTM(100,activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(1))

# Printing the model summary
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss',
                               patience=5,
                               restore_best_weights=True)
```

```python
history = model.fit(X_train, y_train,
                    epochs=50,
                    batch_size=8,
                    validation_data=(X_val, y_val),
                    callbacks=[early_stopping],
                    verbose=1)
```
